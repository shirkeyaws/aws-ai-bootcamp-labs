```python
import boto3
import IPython
from pprint import pprint

client = boto3.client('translate', region_name='us-east-1')
polly = boto3.client('polly', region_name='us-east-1')
```

```python
event_name = 'Amazon Web Services Summit Singapore'
message = 'Hello everyone, welcome to the {}. We hope you enjoy the day!'.format(event_name)

result=client.translate_text(
    Text=message,
    SourceLanguageCode='en',
    TargetLanguageCode='de'
)
pprint(result)
```

```python
response = polly.synthesize_speech(
    Text=result['TranslatedText'],
    TextType="text",
    OutputFormat="mp3",                                           
    VoiceId="Hans")

pprint (response)


outfile = "pollyresponse_de.mp3"
data = response['AudioStream'].read()

with open(outfile,'wb') as f:
     f.write(data)
IPython.display.Audio(outfile) 
```
